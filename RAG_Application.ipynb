```python
from langchain.document_loaders import TextLoader
from langchain.vectorstores import FAISS
```

```python
with open("/content/drive/MyDrive/state_of_the_union.txt","r", encoding="utf8") as f:
  data=f.read()
```

```python
loader=TextLoader("/content/drive/MyDrive/state_of_the_union.txt")
```

```python
document=loader.load()
```

```python
print(document[0].page_content)
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
```

```python
text_chunks=text_splitter.split_documents(document)
```

```python
import os
os.environ["GOOGLE_API_KEY"] = "AIzaSyBa3AS6XHcQmVdyfcYRIhJg5fNro2jQq1c"

from langchain_google_genai import GoogleGenerativeAIEmbeddings

embeddings = GoogleGenerativeAIEmbeddings(
    model = "models/gemini-embedding-001"
)
```

```python
!pip install faiss-cpu
```

```python
vectorstore = FAISS.from_documents(text_chunks, embeddings)
```

```python
retriever = vectorstore.as_retriever()
```

```python
from langchain.prompts import ChatPromptTemplate
```

```python
template = """ You are a smart assistant for question-answering tasks.
Use the following pieces of retrieved context to answer the question.
If you don't know the answer, just say that you don't know.
Use 20 sentences maximum and keep the answer concise.
Question: {question}
Context: {context}
Helpful Answer:
"""
```

```python
prompt=ChatPromptTemplate.from_template(template)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
output_parser = StrOutputParser()
```

```python
import os
os.environ["GOOGLE_API_KEY"] = "AIzaSyBa3AS6XHcQmVdyfcYRIhJg5fNro2jQq1c"
llm = ChatGoogleGenerativeAI(
    model = "gemini-2.5-pro"
)
```

```python
reg_chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | llm
    | output_parser
)
```

```python
reg_chain.invoke("What did the president say about Ketanji Brown Jackson?")
```
